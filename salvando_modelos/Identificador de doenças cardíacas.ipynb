```python
import pickle

modelo = open('modelo_cardiac_diseases','rb')
lm_new = pickle.load(modelo)
modelo.close()
```

```python
print("IDENTIFICADOR E CLASSIFICADOR DE DOENÇAS CARDÍACAS")
print("=-"*30)
while True:
    try:
        age = int(input('Idade:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a idade.')
while True:
    try:
        sex = int(input('Sexo(0: Masculino, 1: Feminino):'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o sexo.')
    except (sex > 1 or sex < 0):
        print('Por Favor, insira um valor válido para o sexo.')
while True:
    try:
        weight = float(input('Peso:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o peso.')
while True:
    try:
        height = float(input('Altura:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a altura.')
while True:
    try:
        qrs_duration = float(input('Duração qrs:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a duração qrs.')
while True:
    try:  
        p_r_interval = float(input('Intervalo pr:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo pr.')
while True:
    try:
        q_t_interval = float(input('Intervalo qt:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo qt.')
while True:
    try:
        t_interval = float(input('Intervalo t:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo t.')
while True:
    try:
        p_interval = float(input('Intervalo p:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo p.')
while True:
    try:
        qrs = float(input('qrs:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo qrs.')
while True:
    try:        
        T = float(input('T:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo T')
while True:
    try: 
        P = float(input('P:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo P')
while True:
    try:
        QRST = float(input('QRST:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para o intervalo QRST')
while True:
    try:
        heart_rate = float(input('Frequência Cardíaca:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a frequência cardíaca')
while True:
    try:        
        q_wave = float(input('Onda q:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a onda q')
while True:
    try:       
        r_wave = float(input('Onda r:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a onda r')
while True:
    try:
        s_wave = float(input('Onda s:'))
        break
    except ValueError:
        print('Por Favor, insira um valor válido para a onda s')
while True:
    try:        
        R_wave = float(input("Onda R':"))
        break
    except ValueError:
        print("Por Favor, insira um valor válido para a onda R'")
while True:
    try:
        S_wave = float(input("Onda S':"))
        break
    except ValueError:
        print("Por Favor, insira um valor válido para a onda S'")
variaveis = [age, sex, weight, height, qrs_duration, p_r_interval, q_t_interval, t_interval, p_interval,
          qrs, T, P, QRST, heart_rate, q_wave, r_wave, s_wave, R_wave, S_wave]

previsao = lm_new.predict([variaveis])
previsao

codigo = {1:'Normal',
   2:'Ischemic changes (Coronary Artery Disease)',
   3:'Old Anterior Myocardial Infarction ',
   4:'Old Inferior Myocardial Infarction', 
   5:'Sinus tachycardy',
   6:'Sinus bradycardy',  
   7:'Ventricular Premature Contraction (PVC)', 
   8:'Supraventricular Premature Contraction', 
   9:'Left bundle branch block',  
   10:'Right bundle branch block',
   11:'1. degree AtrioVentricular block',   
   12:'2. degree AV block',
   13:'3. degree AV block',
   14:'Left ventricule hypertrophy',
   15:'Atrial Fibrillation or Flutter',
   16:'Others'}

codigo[previsao[0]]
```
